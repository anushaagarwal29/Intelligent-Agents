```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('FinalResults-Additional.csv', header = None)
```

```python
df
```

```python
average_values = df.groupby(0)[1].mean()
```

```python
# Sort the average values in descending order
average_values_sorted = average_values.sort_values(ascending=False)
```

```python
print(average_values_sorted)
```

```python
# Plot the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=average_values_sorted.index, y=average_values_sorted.values, palette='viridis')

# Annotate scores on top of the bars
for i, v in enumerate(average_values_sorted.values):
    plt.text(i, v + 1, str(round(v, 2)), ha='center', va='bottom', fontsize=9)

plt.title('Average Scores Across 50 Tournaments for PD Strategies')
plt.xlabel('PD Strategies')
plt.ylabel('Average Score')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
