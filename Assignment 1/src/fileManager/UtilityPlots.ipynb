```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plotUtility(df):
    # Extract headers and utility estimates
    headers = df.columns.tolist()
    iterations = range(1, len(df) + 1)

    # Plot utility estimates across iterations for each header value
    plt.figure(figsize=(8, 4))
    for header in headers:
        plt.plot(iterations, df[header], label=header)

    # Add labels and legend
    plt.xlabel('Iteration')
    plt.ylabel('Utility Estimate')
    plt.title('Utility Estimates Across Iterations')
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), fancybox=True, shadow=True, ncol=6)
    plt.grid(True)

    # Show plot
    plt.show()
```

```python
import os

def process_files(folder_path):
    """
    Process all files in a folder and its subfolders.
    
    Parameters:
        folder_path (str): Path to the base folder.
    """
    # Iterate over all items in the folder
    for item in os.listdir(folder_path):
        item_path = os.path.join(folder_path, item)
        if os.path.isfile(item_path):
            # Process the file (replace this with your function)
            print(item_path)
            df = pd.read_csv(item_path)
            df = df.iloc[:, :-1]
            plotUtility(df);
            
        elif os.path.isdir(item_path):
            # Recursively process files in subfolders
            process_files(item_path)
```
